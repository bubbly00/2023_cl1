```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("wine-quality-white-and-red-2.csv")

# Separate features (X) from the target variable (wine type)
X = data.drop('type', axis=1)
y = data['type']

# Standardize the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Apply PCA with 2 components for visualization
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_std)

# Create a DataFrame with the first two principal components and the target variable
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2','Principal Component 3'])
df_pca['type'] = y

# Visualize the data
plt.figure(figsize=(10, 6))
for wine_style in df_pca['type'].unique():
    plt.scatter(
        df_pca[df_pca['type'] == wine_style]['Principal Component 1'],
        df_pca[df_pca['type'] == wine_style]['Principal Component 2'],
        df_pca[df_pca['type'] == wine_style]['Principal Component 3'],
        label=f'{wine_style}',
        alpha=0.7
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.ylabel('Principal Component 3')
plt.title('PCA - Wine Dataset')
plt.legend()
plt.grid(True)
plt.show()
```
