## Problem Statement

Implement Agglomerative hierarchical clustering algorithm using appropriate dataset. ## Problem Statement

### Background

Hierarchical clustering is a widely-used method for unsupervised data analysis. It creates a hierarchy of clusters, making it easy to visualize how data points are grouped together. Dendrograms are commonly used to display these hierarchical structures.


### Input

The input for this problem is a sample dataset, which consists of data points in two dimensions. The dataset is generated using the `make_blobs` function.

### Output

The output is a clear and informative dendrogram that visualizes the hierarchical clustering structure of the sample dataset. The dendrogram should help identify clusters and relationships among data points.

### Implementation

The provided code accomplishes the hierarchical clustering and dendrogram generation tasks. By adjusting parameters such as the linkage method and dendrogram simplification, the code aims to create a dendrogram that aids in data analysis and clustering understanding.

### Potential Improvements

For larger or more complex datasets, further customization of the dendrogram, including color-coding, labeling, and interactive elements, can enhance the interpretation of the hierarchical clustering structure. Additionally, real-world datasets can be used for more practical applications of hierarchical clustering and dendrogram visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
```

### Generating a sample dataset

```python
# Generating a sample dataset
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)
```

### Performing Agglomerative Hierarchical Clustering

```python
n_clusters = 3
agg_cluster = AgglomerativeClustering(n_clusters=n_clusters)
agg_cluster.fit(X)
```

```python
# Performing hierarchical clustering with a different linkage method (e.g., 'average')
linked = linkage(X, method='average')
dendrogram(linked, truncate_mode='level', p=3)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
#  Plotting the clustering results
plt.scatter(X[:, 0], X[:, 1], c=agg_cluster.labels_, cmap='rainbow')
plt.title("Agglomerative Hierarchical Clustering")
plt.show()
```
